# 13. 이미지 변형 (이진화)

## Threshold

```python
import cv2

img = cv2.imread('book.jpg', cv2.IMREAD_GRAYSCALE)

ret, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
# 역치값을 127 로 잡고 그보다 넘었을시 255(흰색) 처리 그게 아니면 검정색 처리 하기


cv2.imshow('img',img)
cv2.imshow('binary',binary)

cv2.waitKey(0)
cv2.destroyAllWindows()

```

## Trackbar (값 변화에 따른 변형 확인)

```python
import cv2

def empty(pos):
    print(pos) # track bar 의 위치
    pass


img = cv2.imread("book.jpg",cv2.IMREAD_GRAYSCALE)

name = 'Trackbar'
cv2.namedWindow(name)

# 0 : 검은색, 127 : 진한 회색, 195 : 밝은 회색, 255 : 흰색
cv2.createTrackbar('threshold',name, 127, 255, empty) # bar의 이름, 창의이름, 초기값, 최대값, 이벤트 처리

while True:
    thresh = cv2.getTrackbarPos('threshold',name) # bar 이름, 창의 이름
    ret, binary = cv2.threshold(img,thresh,255,cv2.THRESH_BINARY)
    
    if not ret:
        break
    
    cv2.imshow(name,binary)
    if cv2.waitKey(1) == ord('q'):
        break


cv2.destroyAllWindows()
```

## Adaptive Threshold
이미지 영역을 세분화해서 잘게짤라서 임계치를 적용하는 방법

```python
import cv2

def empty(pos):
    pass

img = cv2.imread('book.jpg',cv2.IMREAD_GRAYSCALE)
name = 'Trackbar'
cv2.namedWindow(name)

# bar 이름, 창의이름, 초기값, 최대값, 이벤트 처리
cv2.createTrackbar('c',name,3,10,empty) # 일반적으로 양수의 값을 사용
cv2.createTrackbar('block_size',name,25,100,empty) # 블록사이즈는 홀수만가능 ,1보다는 큰값


while True:
    block_size = cv2.getTrackbarPos("block_size",name) # bar 이름, 창의 이름
    c = cv2.getTrackbarPos('c',name)
    
    if block_size <=1: # 1 이하면 3으로
        block_size = 3
        
    if block_size %2==0: # 짝수이면 홀수로
        block_size += 1
    
    # 이미지, 임계치를 넘었을때 처리해주는 값, 
    binary = cv2.adaptiveThreshold(img ,255 ,cv2.ADAPTIVE_THRESH_MEAN_C ,cv2.THRESH_BINARY ,block_size ,c)
    
    cv2.imshow(name,binary)    
    
    if cv2.waitKey(0) == ord('q'):
        break
        
cv2.destroyAllWindows()
```

## 오츠 알고리즘
threshold 를 이용해서 바꾸면서 확인하기 불편하니까 최적의 임계치값을 찾는걸 함
Bimodal Image 에 사용하기 적합나다( 최적의 임계치를 자동으로 발견)

```python
import cv2

img = cv2.imread('book.jpg',cv2.IMREAD_GRAYSCALE)

ret, binary = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret, otsu = cv2.threshold(img, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
print("otsu threshold",ret)


cv2.imshow('img',img)
cv2.imshow('binary',binary)
cv2.imshow("otsu",otsu)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
